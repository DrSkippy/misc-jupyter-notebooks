```python
import math

rnd = 3
```

```python
def salary_in_year(n, alpha=0.025, delay=0):
    return (1 + alpha) ** max(0, n - delay)


def buying_power_in_year(n, infl=0.045):
    return (1 - infl) ** n


def sum_salary(n, alpha=0.025, infl=0.045, delay=0):
    s = [salary_in_year(i, alpha, delay) for i in range(n)]
    si = [buying_power_in_year(i, infl) * salary_in_year(i, alpha, delay) for i in range(n)]
    print("*" * 60)
    print("n={} years alpha={}% inflation={}% delay={} years".format(n, alpha * 100, infl * 100, delay))
    return s, si, sum(s), sum(si), salary_in_year(n - 1, alpha, delay), (buying_power_in_year(n - 1, infl) *
                                                                         salary_in_year(n - 1, alpha, delay))


def show(s, si, sum_s, sum_si, sal_final, bp_final):
    print("Salary in year:\n  ", "\n   ".join([str(i) + ": " + str(round(x, rnd)) for i, x in enumerate(s)]))
    print("Buying power in year:\n  ", "\n   ".join([str(i) + ": " + str(round(x, rnd)) for i, x in enumerate(si)]))
    print("Sum salary: ", round(sum_s, rnd))
    print("Sum buying power: ", round(sum_si, rnd))
    print("Salary final: ", round(sal_final, rnd))
    print("Buying power final: ", round(bp_final, rnd))


def diff(a, b):
    d1 = [a[0][i] - b[0][i] for i in range(len(a[0]))]
    d2 = [a[1][i] - b[1][i] for i in range(len(a[1]))]
    d3 = [a[i] - b[i] for i in range(2, len(a))]
    return d1, d2, *d3
```

```python
show(*sum_salary(n=10, alpha=0.025, infl=0.0, delay=0))
show(*sum_salary(n=10, delay=0))
show(*sum_salary(n=10, delay=1))

a = sum_salary(n=10, alpha=0.025, infl=0.025, delay=0)
show(*a)
b = sum_salary(n=10, alpha=0.025, infl=0.025, delay=1)
show(*b)
print("*"*60)
print("Comparison...")
c = diff(b, a)
show(*c)
```
